```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data=pd.read_csv("sonardata.csv",header=None)
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[10].value_counts()
```

```python
sonar_data.groupby(10).mean()
```

```python
X=sonar_data.drop(columns=10,axis=1)
y=sonar_data[10]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_pred=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_pred,Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_pred=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_pred,Y_test)
```

```python
print(test_data_accuracy)
```

```python
input_data_mine = (0.0412, 0.1135, 0.0518, 0.0232, 0.0646, 0.1124, 0.1787, 0.2407, 0.2682, 0.2058)
input_data_rock = (0.0362, 0.0192, 0.0167, 0.0368, 0.0829, 0.1347, 0.1478, 0.0902, 0.1305, 0.3526)

mine_samples = [input_data_mine] * 10
rock_samples = [input_data_rock] * 10

mine_samples_as_numpy_array = np.asarray(mine_samples)
rock_samples_as_numpy_array = np.asarray(rock_samples)

predictions_mine = model.predict(mine_samples_as_numpy_array)
predictions_rock = model.predict(rock_samples_as_numpy_array)

print("Predictions for 'Mine' data:")
for prediction in predictions_mine:
    if prediction == "M":
        print("Input data for 'Mine' is classified as Mine!")
    else:
        print("Input data for 'Mine' is not classified as Mine.")

print("\nPredictions for 'Rock' data:")
for prediction in predictions_rock:
    if prediction == "R":
        print("Input data for 'Rock' is classified as Rock!")
    else:
        print("Input data for 'Rock' is not classified as Rock.")

```
